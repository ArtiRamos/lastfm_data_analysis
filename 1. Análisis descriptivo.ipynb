¿Cuál es el Top 10 de canciones en la lista Top Tracks y la proporción que representa?

```python
df_tracks_sorted = df_tracks.sort_values(by = "playcount", ascending = False)
df_tracks_sorted.reset_index()
```

```python
top_10_tracks = df_tracks_sorted.head(10)
top_10_tracks.reset_index()
```

```python
top_10_tracks_playcount = top_10_tracks["playcount"].sum()
total_tracks_playcount = df_tracks["playcount"].sum()
proportion_top_10_tracks = (top_10_tracks_playcount / total_tracks_playcount)*100
print(f"La proporción que representa el Top 10 canciones es {proportion_top_10_tracks:.2f}%")
```
