```python
#Ahora vamos a sacar el numero total de canciones
total_tracks= len(dfy)
```

```python
#Aquí queremos sacar el número que representa el Top 10%
top_10_Tracks_count = int(total_tracks * 0.10)
```

```python
#Así que ordenamos por la columna Playcount
dfy_sorted = dfy.sort_values(
    by = "playcount",
    ascending = False
)
```

```python
# Y seleccionamos ese Top 10
top_10_tracks = dfy_sorted.head(top_10_Tracks_count)
```

```python
# y aquí calculamos la proporción
top_10_tracks_playcount = top_10_tracks["playcount"].sum()
total_tracks_playcount = dfy["playcount"].sum()
proportion_top_10_tracks = top_10_tracks_playcount / total_tracks_playcount
```

```python
# Y finalmente podemos ver que el top 10% de Artistas, concentra el 31% de todas las canciones incluidas en la tabla Top Tracks
proportion_top_10_tracks
```
