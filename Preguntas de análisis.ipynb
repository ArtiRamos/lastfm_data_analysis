```python
import requests
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_artist = pd.read_csv("top_artist_lastfm.csv")
```

```python
df_artist.info()
```

```python
df_artist.head(3)
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df_artist["playcount"], bins=30)
plt.xlabel("Playcount")
plt.ylabel("Número de artistas")
plt.title("Distribución del playcount de artistas")
plt.show()
```

```python
#12 
Distribución sesgada completamente a la derecha. 
    Casi logarítmica.
```

```python
df_tracks = pd.read_csv("top_tracks_lastfm.csv")
```

```python
df_tracks.head(3)
```

```python
#13 DISTRIBUCIÓN DEL PLAYCOUNT DE TRACKS

plt.figure(figsize=(8, 5))
plt.hist(df_tracks["playcount"], bins=30)
plt.xlabel("Playcount")
plt.ylabel("Número de canción")
plt.title("Distribución del playcount de tracks")
plt.show()
```

#14 lONG TAIL EN CANCIONES

Ambas distribuciones se consideran de Long Tail. En ellas pocos elementos concentran
la mayoría del valor y el resto son muchos elementos con valores muy pequeños.

```python
#15 GRÁFICO TOP 10 ARTISTAS POR PLAYCOUNT
df_artist_top10 = df_artist.sort_values("playcount", ascending = False).head(10)
```

```python
df_artist_top10
```

```python
sns.barplot(data = df_artist_top10, x = "playcount", y = "name", color = "green")
```

```python
#16 GRÁFICO DE TOP 10 CANCIONES POR PLAYCOUNT
df_tracks_top10 = df_tracks.sort_values("playcount", ascending = False).head(10)
```

```python
df_tracks_top10
```

```python
sns.barplot(data = df_tracks_top10, x = "playcount", y = "name", color = "magenta")
```

```python
#17 DISTRIBUCIÓN DE LISTENERS POR ARTISTA
plt.figure(figsize=(8, 5))
plt.hist(df_tracks["listeners"], bins=30)
plt.xlabel("Listeners")
plt.ylabel("Número de artistas")
plt.title("Distribución de oyentes por artistas")
plt.show()
```

```python
#18. IDENTIFICACIÓN DE OUTLIERS DATA FRAME ARTISTAS
plt.figure(figsize=(6, 5))
plt.boxplot(df_artist["playcount"])
plt.ylabel("Playcount")
plt.title("Outliers en playcount de artistas")
plt.show()
```

```python
18# OUTLIERS DATA FRAME TRACKS
plt.figure(figsize=(6, 5))
plt.boxplot(df_tracks["playcount"])
plt.ylabel("Playcount")
plt.title("Outliers en playcount de tracks")
plt.show()
```

```python
#19. COMPARACIÓN VISUAL ARTISTAS VS TRACKS

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df_artist["playcount"], bins=30)
plt.title("Playcount de artistas")
plt.xlabel("Playcount")
plt.ylabel("Frecuencia")

plt.subplot(1, 2, 2)
plt.hist(df_tracks["playcount"], bins=30)
plt.title("Playcount de tracks")
plt.xlabel("Playcount")
plt.ylabel("Frecuencia")

plt.show()
```

Ambos gráficos nos presenta una distribución de Long Tail. En en DataFrame de tracks observamos mayor dispersión que en la de artistas.
